1. What is hypothesis testing in statistics

     - Hypothesis testing is a statistical method used to make inferences or draw conclusions about a population based on a sample of data. It involves formulating two competing hypotheses and using sample data to determine which hypothesis is more likely to be true.
     - Types of Errors in Hypothesis Testing
         - Type I Error (Œ±)
         - Type II Error (Œ≤)

2. What is the null hypothesis, and how does it differ from the alternative
  hypothesis

     - Null Hypothesis
        - Assumes no change or effect
        - To be tested and possibly rejected
        - The average test score is 75

  - Alternative Hypothesis
     - Suggests a change or effect
     - To be supported if evidence suggests
     - The average test score is not 75.

3. What is the significance level in hypothesis testing, and why is it
  important

    - The significance level in hypothesis testing is the probability of rejecting the null hypothesis when it is actually true (Type I error). It represents the threshold for determining statistical significance.
    - Common Significance Levels:
       - 0.05 (5%) - Most commonly used
       - 0.01 (1%) - More stringent, used in highly critical studies
       - 0.10 (10%) - Sometimes used in exploratory research

4. What does a P-value represent in hypothesis testing

   - The p-value is a fundamental concept in statistical inference that quantifies the probability of obtaining results at least as extreme as the observed data, assuming that the null hypothesis is true.
   - The probability that ùêª0 is true.
   - The magnitude or practical significance of the effect.
   - The probability of making an error directly.

5. How do you interpret the P-value in hypothesis testing

   - Interpreting the P-Value in Hypothesis Testing
   - The p-value represents the probability of obtaining a test statistic as extreme as, or more extreme than, the observed result if the null hypothesis (ùêª0) is true. It helps determine whether to reject ùêª0in favor of the alternative hypothesis (ùêªa).
   

6. What are Type 1 and Type 2 errors in hypothesis testing

   1. Type I Error (False Positive)
       - Occurs when H0(null hypothesis) is true, but we reject it.
       - Probability of Type I error = Œ± (significance level)
       - Example: A medical test wrongly detects a disease in a healthy person.

  2. Type II Error (False Negative)
      - Occurs when ùêª0 is false, but we fail to reject it.
      - Probability of Type II error = Œ≤.
      - Example: A medical test fails to detect a disease in a sick person.

```python
# Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

import scipy.stats as stats
import math

def z_test(sample_mean, population_mean, std_dev, sample_size, alpha=0.05, tail="two"):
    z_score = (sample_mean - population_mean) / (std_dev / math.sqrt(sample_size))


    if tail == "two":
        p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
    elif tail == "left":
        p_value = stats.norm.cdf(z_score)
    elif tail == "right":
        p_value = 1 - stats.norm.cdf(z_score)
    else:
        raise ValueError("Invalid tail type. Choose 'two', 'left', or 'right'.")

    # Interpretation
    print(f"Z-Score: {z_score:.4f}")
    print(f"P-Value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis (H0). There is significant evidence.")
    else:
        print("Fail to reject the null hypothesis (H0). No significant difference.")

sample_mean = 52
population_mean = 50
std_dev = 5
sample_size = 30
alpha = 0.05
tail = "two"

z_test(sample_mean, population_mean, std_dev, sample_size, alpha, tail)
```

```python
# Implement a one-sample Z-test using Python to compare the sample mean with the population mean

import numpy as np
import scipy.stats as stats

def one_sample_z_test(sample, population_mean, population_std_dev, alpha=0.05, tail="two"):
    sample_size = len(sample)
    sample_mean = np.mean(sample)

    # Calculate Z-score
    z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

    # Calculate p-value
    if tail == "two":
        p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test
    elif tail == "left":
        p_value = stats.norm.cdf(z_score)  # Left-tailed test
    elif tail == "right":
        p_value = 1 - stats.norm.cdf(z_score)  # Right-tailed test
    else:
        raise ValueError("Invalid tail type. Choose 'two', 'left', or 'right'.")

    # Print results
    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Z-Score: {z_score:.4f}")
    print(f"P-Value: {p_value:.4f}")

    # Decision Rule
    if p_value < alpha:
        print("Reject the null hypothesis (H0): Significant difference detected.")
    else:
        print("Fail to reject the null hypothesis (H0): No significant difference.")

# Simulated sample data (normally distributed)
np.random.seed(42)
sample_data = np.random.normal(loc=52, scale=10, size=30)  # Sample mean ‚â† Population mean

# Define known population parameters
population_mean = 50
population_std_dev = 10  # Known population standard deviation

# Perform the Z-test (two-tailed)
one_sample_z_test(sample_data, population_mean, population_std_dev, alpha=0.05, tail="two")
```

```python
# Perform a two-tailed Z-test using Python and visualize the decision region on a plot

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def two_tailed_z_test(sample, population_mean, population_std_dev, alpha=0.05):
    sample_size = len(sample)
    sample_mean = np.mean(sample)

    # Calculate Z-score
    z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

    # Calculate p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    # Critical value (for two-tailed test)
    critical_value = stats.norm.ppf(1 - alpha / 2)

    # Print results
    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Z-Score: {z_score:.4f}")
    print(f"P-Value: {p_value:.4f}")
    print(f"Critical Z-Value: ¬±{critical_value:.4f}")

    # Decision Rule
    if abs(z_score) > critical_value:
        print("Reject the null hypothesis (H0): Significant difference detected.")
    else:
        print("Fail to reject the null hypothesis (H0): No significant difference.")

    # Visualization
    plot_z_test(z_score, critical_value)

def plot_z_test(z_score, critical_value):
    # Generate values for the normal distribution
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x, 0, 1)

    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label="Standard Normal Distribution", color="blue")

    # Shade the rejection regions
    plt.fill_between(x, y, where=(x <= -critical_value) | (x >= critical_value), color='red', alpha=0.3, label="Rejection Region")

    # Plot the Z-score
    plt.axvline(z_score, color='black', linestyle='dashed', label=f"Z-Score = {z_score:.4f}")

    # Add labels
    plt.axvline(-critical_value, color='red', linestyle='dashed', label=f"Critical Value = ¬±{critical_value:.4f}")
    plt.axvline(critical_value, color='red', linestyle='dashed')

    plt.title("Two-Tailed Z-Test Visualization")
    plt.xlabel("Z-Score")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid()
    plt.show()

# Simulated sample data (normally distributed)
np.random.seed(42)
sample_data = np.random.normal(loc=52, scale=10, size=30)  # Sample with mean ‚â† Population mean

# Define known population parameters
population_mean = 50
population_std_dev = 10  # Known population standard deviation

# Perform the two-tailed Z-test
two_tailed_z_test(sample_data, population_mean, population_std_dev, alpha=0.05)
```

```python
# Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_type1_type2_errors(mu_null=50, mu_alt=53, sigma=10, sample_size=30, alpha=0.05):
    """
    Visualizes Type 1 and Type 2 errors in hypothesis testing.

    Parameters:
    - mu_null: Mean under H0
    - mu_alt: Mean under H1
    - sigma: Population standard deviation
    - sample_size: Size of the sample
    - alpha: Significance level (default = 0.05)
    """
    # Calculate standard error
    standard_error = sigma / np.sqrt(sample_size)

    # Compute critical value (for two-tailed test)
    z_critical = stats.norm.ppf(1 - alpha / 2)  # Two-tailed test
    critical_value = mu_null + z_critical * standard_error  # Right-tail critical value

    # Compute Type 2 error (Beta)
    beta = stats.norm.cdf((critical_value - mu_alt) / standard_error)
    power = 1 - beta  # Power of the test

    # Generate x values for the normal curves
    x = np.linspace(mu_null - 4 * standard_error, mu_alt + 4 * standard_error, 1000)
    y_null = stats.norm.pdf(x, mu_null, standard_error)
    y_alt = stats.norm.pdf(x, mu_alt, standard_error)

    # Plot the distributions
    plt.figure(figsize=(10, 5))
    plt.plot(x, y_null, label="H0: Null Distribution", color="blue")
    plt.plot(x, y_alt, label="H1: Alternative Distribution", color="green")

    # Shade Type 1 Error (Alpha)
    plt.fill_between(x, y_null, where=(x >= critical_value), color='red', alpha=0.3, label="Type 1 Error (Œ±)")

    # Shade Type 2 Error (Beta)
    plt.fill_between(x, y_alt, where=(x < critical_value), color='purple', alpha=0.3, label="Type 2 Error (Œ≤)")

    # Mark critical value
    plt.axvline(critical_value, color='black', linestyle='dashed', label=f"Critical Value = {critical_value:.2f}")

    # Labels & Legends
    plt.title("Type 1 & Type 2 Errors in Hypothesis Testing")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid()
    plt.show()

    # Print results
    print(f"Critical Value: {critical_value:.2f}")
    print(f"Type 1 Error (Alpha): {alpha}")
    print(f"Type 2 Error (Beta): {beta:.4f}")
    print(f"Power of the Test (1 - Beta): {power:.4f}")

# Run the function
visualize_type1_type2_errors(mu_null=50, mu_alt=53, sigma=10, sample_size=30, alpha=0.05)
```

```python
# Write a Python program to perform an independent T-test and interpret the results

import numpy as np
import scipy.stats as stats

def independent_t_test(sample1, sample2, alpha=0.05, equal_var=True):
    """
    Performs an independent two-sample T-test.

    Parameters:
    - sample1: First sample (array-like)
    - sample2: Second sample (array-like)
    - alpha: Significance level (default = 0.05)
    - equal_var: Assumption of equal variances (default = True)

    Returns:
    - T-statistic, P-value, and interpretation
    """
    # Perform independent T-test
    t_stat, p_value = stats.ttest_ind(sample1, sample2, equal_var=equal_var)

    # Print results
    print(f"Sample 1 Mean: {np.mean(sample1):.4f}")
    print(f"Sample 2 Mean: {np.mean(sample2):.4f}")
    print(f"T-Statistic: {t_stat:.4f}")
    print(f"P-Value: {p_value:.4f}")

    # Decision Rule
    if p_value < alpha:
        print("Reject the null hypothesis (H0): Significant difference detected.")
    else:
        print("Fail to reject the null hypothesis (H0): No significant difference.")

# Simulate two independent samples (normally distributed)
np.random.seed(42)
sample1 = np.random.normal(loc=55, scale=10, size=30)  # Mean = 55, Std Dev = 10
sample2 = np.random.normal(loc=50, scale=10, size=30)  # Mean = 50, Std Dev = 10

# Perform the independent T-test
independent_t_test(sample1, sample2, alpha=0.05, equal_var=True)
```

```python
# Perform a paired sample T-test using Python and visualize the comparison results

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def paired_t_test(before, after, alpha=0.05):
    """
    Performs a paired (dependent) T-test and visualizes the results.

    Parameters:
    - before: Data before treatment (array-like)
    - after: Data after treatment (array-like)
    - alpha: Significance level (default = 0.05)
    """
    # Perform Paired T-test
    t_stat, p_value = stats.ttest_rel(before, after)

    # Print results
    print(f"Mean Before: {np.mean(before):.4f}")
    print(f"Mean After: {np.mean(after):.4f}")
    print(f"T-Statistic: {t_stat:.4f}")
    print(f"P-Value: {p_value:.4f}")

    # Decision Rule
    if p_value < alpha:
        print("Reject the null hypothesis (H0): Significant difference detected.")
    else:
        print("Fail to reject the null hypothesis (H0): No significant difference.")

    # Visualization
    plot_paired_t_test(before, after)

def plot_paired_t_test(before, after):
    """
    Plots the before and after results for visual comparison.
    """
    plt.figure(figsize=(8, 5))

    # Create scatter plot
    plt.scatter(np.arange(len(before)), before, color='blue', label="Before", alpha=0.7)
    plt.scatter(np.arange(len(after)), after, color='green', label="After", alpha=0.7)

    # Draw lines connecting paired values
    for i in range(len(before)):
        plt.plot([i, i], [before[i], after[i]], color='gray', linestyle='dotted')

    # Labels & Titles
    plt.title("Paired Sample T-Test Comparison")
    plt.xlabel("Subject Index")
    plt.ylabel("Measurement Value")
    plt.legend()
    plt.grid()
    plt.show()

# Simulated data: Scores before & after training
np.random.seed(42)
before_scores = np.random.normal(loc=50, scale=10, size=30)  # Before treatment
after_scores = before_scores + np.random.normal(loc=3, scale=5, size=30)  # After treatment

# Perform Paired T-test
paired_t_test(before_scores, after_scores, alpha=0.05)
```

```python
# Simulate data and perform both Z-test and T-test, then compare the results using Python

import numpy as np
import scipy.stats as stats

def z_test(sample, pop_mean, pop_std, alpha=0.05):
    """
    Performs a one-sample Z-test.

    Parameters:
    - sample: Sample data (array-like)
    - pop_mean: Population mean
    - pop_std: Population standard deviation
    - alpha: Significance level (default = 0.05)
    """
    sample_mean = np.mean(sample)
    sample_size = len(sample)
    standard_error = pop_std / np.sqrt(sample_size)

    # Compute Z-score
    z_score = (sample_mean - pop_mean) / standard_error
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test

    # Print results
    print("\n=== Z-Test Results ===")
    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Z-Score: {z_score:.4f}")
    print(f"P-Value: {p_value:.4f}")

    # Decision Rule
    if p_value < alpha:
        print("Reject the null hypothesis (H0): Significant difference detected.")
    else:
        print("Fail to reject the null hypothesis (H0): No significant difference.")

def t_test(sample, pop_mean, alpha=0.05):
    """
    Performs a one-sample T-test.

    Parameters:
    - sample: Sample data (array-like)
    - pop_mean: Population mean
    - alpha: Significance level (default = 0.05)
    """
    t_stat, p_value = stats.ttest_1samp(sample, pop_mean)

    # Print results
    print("\n=== T-Test Results ===")
    print(f"Sample Mean: {np.mean(sample):.4f}")
    print(f"T-Statistic: {t_stat:.4f}")
    print(f"P-Value: {p_value:.4f}")

    # Decision Rule
    if p_value < alpha:
        print("Reject the null hypothesis (H0): Significant difference detected.")
    else:
        print("Fail to reject the null hypothesis (H0): No significant difference.")

# Simulating random sample data
np.random.seed(42)
sample_size = 30  # Small sample size for T-test
pop_mean = 50     # Population mean
pop_std = 10      # Known population standard deviation

# Generate sample data (normally distributed)
sample_data = np.random.normal(loc=52, scale=pop_std, size=sample_size)

# Perform both tests
z_test(sample_data, pop_mean, pop_std, alpha=0.05)
t_test(sample_data, pop_mean, alpha=0.05)
```

```python
# Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
import scipy.stats as stats

def confidence_interval(sample, confidence=0.95):
    """
    Calculates the confidence interval for a sample mean.

    Parameters:
    - sample: Sample data (array-like)
    - confidence: Confidence level (default = 0.95)

    Returns:
    - Lower bound, Upper bound of the confidence interval
    """
    sample_mean = np.mean(sample)  # Sample mean
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation (ddof=1 for sample)
    n = len(sample)  # Sample size

    # Compute margin of error using t-distribution (for small sample sizes)
    t_score = stats.t.ppf((1 + confidence) / 2, df=n-1)  # Two-tailed test
    margin_of_error = t_score * (sample_std / np.sqrt(n))

    # Confidence interval bounds
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Simulated Sample Data
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)  # Sample with mean=50, std=10

# Calculate 95% Confidence Interval
ci_lower, ci_upper = confidence_interval(sample_data, confidence=0.95)

# Print the result
print(f"95% Confidence Interval: ({ci_lower:.4f}, {ci_upper:.4f})")
```

```python
# Write a Python program to calculate the margin of error for a given confidence level using sample data

import numpy as np
import scipy.stats as stats

def margin_of_error(sample, confidence=0.95):
    """
    Calculates the margin of error for a given confidence level.

    Parameters:
    - sample: Sample data (array-like)
    - confidence: Confidence level (default = 0.95)

    Returns:
    - Margin of error
    """
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation (ddof=1 for sample)
    n = len(sample)  # Sample size

    # Get the critical value from the t-distribution
    t_score = stats.t.ppf((1 + confidence) / 2, df=n-1)

    # Compute margin of error
    moe = t_score * (sample_std / np.sqrt(n))

    return moe

# Simulated Sample Data
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)  # Sample with mean=50, std=10

# Calculate Margin of Error for 95% Confidence Level
moe_95 = margin_of_error(sample_data, confidence=0.95)

# Print the result
print(f"Margin of Error (95% Confidence Level): {moe_95:.4f}")
```

```python
# Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

def bayes_theorem(prior_A, likelihood_B_given_A, likelihood_B_given_not_A):
    """
    Computes the posterior probability using Bayes' Theorem.

    Parameters:
    - prior_A: P(A) - Prior probability of event A
    - likelihood_B_given_A: P(B | A) - Likelihood of B given A
    - likelihood_B_given_not_A: P(B | ¬¨A) - Likelihood of B given not A

    Returns:
    - Posterior probability P(A | B)
    """
    # Total probability of B
    prior_not_A = 1 - prior_A  # Complement of prior A
    P_B = (likelihood_B_given_A * prior_A) + (likelihood_B_given_not_A * prior_not_A)

    # Apply Bayes' Theorem
    posterior_A_given_B = (likelihood_B_given_A * prior_A) / P_B

    return posterior_A_given_B

# Example Scenario: Disease Testing
prior_disease = 0.01  # 1% of the population has the disease
test_sensitivity = 0.95  # P(Positive Test | Disease) = 95% (True Positive Rate)
test_false_positive = 0.05  # P(Positive Test | No Disease) = 5% (False Positive Rate)

# Compute posterior probability of having the disease given a positive test
posterior_disease_given_positive = bayes_theorem(prior_disease, test_sensitivity, test_false_positive)

# Print results
print(f"Probability of having the disease given a positive test result: {posterior_disease_given_positive:.4f}")
```

```python
# Implement an F-test using Python to compare the variances of two random samples

import numpy as np
import scipy.stats as stats

# Generate two random samples
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10
sample2 = np.random.normal(loc=50, scale=15, size=30)  # Mean=50, Std=15

# Compute sample variances
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# Compute F-statistic
F_stat = var1 / var2 if var1 > var2 else var2 / var1  # Ensure F > 1

# Degrees of freedom
df1 = len(sample1) - 1
df2 = len(sample2) - 1

# Compute p-value
p_value = 2 * (1 - stats.f.cdf(F_stat, df1, df2))  # Two-tailed test

# Print results
print("=== F-Test for Equality of Variances ===")
print(f"Sample 1 Variance: {var1:.4f}")
print(f"Sample 2 Variance: {var2:.4f}")
print(f"F-Statistic: {F_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Decision Rule at Œ± = 0.05
alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis (H0): The variances are significantly different.")
else:
    print("\nFail to reject the null hypothesis (H0): No significant difference in variances.")
```

```python
# Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results

import numpy as np
import pandas as pd
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Generate sample data
np.random.seed(42)
size = 30  # Number of observations per group

# Factor 1: Treatment (A1, A2)
# Factor 2: Diet (B1, B2)
A = np.repeat(["A1", "A2"], size * 2)
B = np.tile(["B1", "B2"], size * 2)

# Response Variable: Random values with different means
values = np.concatenate([
    np.random.normal(loc=50, scale=10, size=size),  # A1B1
    np.random.normal(loc=55, scale=10, size=size),  # A1B2
    np.random.normal(loc=52, scale=10, size=size),  # A2B1
    np.random.normal(loc=60, scale=10, size=size)   # A2B2
])

# Create DataFrame
df = pd.DataFrame({"Treatment": A, "Diet": B, "Value": values})

# Fit Two-Way ANOVA model
model = ols("Value ~ C(Treatment) + C(Diet) + C(Treatment):C(Diet)", data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Print ANOVA results
print("=== Two-Way ANOVA Test Results ===")
print(anova_table)

# Decision rule (Œ± = 0.05)
alpha = 0.05
for factor in anova_table.index:
    p_value = anova_table.loc[factor, "PR(>F)"]
    if p_value < alpha:
        print(f"\nReject H0 for {factor}: Significant effect found.")
    else:
        print(f"\nFail to reject H0 for {factor}: No significant effect.")

# Visualization: Interaction Plot
plt.figure(figsize=(8, 6))
sns.pointplot(data=df, x="Treatment", y="Value", hue="Diet", palette="Set1", capsize=0.1)
plt.title("Two-Way ANOVA: Interaction Effect of Treatment & Diet")
plt.xlabel("Treatment")
plt.ylabel("Mean Value")
plt.grid(True)
plt.show()
```

```python
# Perform a hypothesis test for population variance using a Chi-square distribution and interpret the resultsD

import numpy as np
import scipy.stats as stats

# Sample data (randomly generated)
np.random.seed(42)
sample = np.random.normal(loc=50, scale=5, size=30)  # Mean=50, Std=5, Size=30

# Given hypothesized population variance
sigma0_squared = 25  # Hypothesized variance (œÉ0^2 = 5^2)

# Sample statistics
n = len(sample)  # Sample size
sample_variance = np.var(sample, ddof=1)  # Sample variance (unbiased estimator)

# Chi-Square Test Statistic
chi_square_stat = (n - 1) * sample_variance / sigma0_squared

# Degrees of freedom
df = n - 1

# Compute p-value (two-tailed test)
p_value = 2 * min(stats.chi2.cdf(chi_square_stat, df), 1 - stats.chi2.cdf(chi_square_stat, df))

# Print Results
print("=== Chi-Square Test for Population Variance ===")
print(f"Sample Variance: {sample_variance:.4f}")
print(f"Chi-Square Statistic: {chi_square_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Decision Rule (Œ± = 0.05)
alpha = 0.05
if p_value < alpha:
    print("\nReject H0: The population variance is significantly different from the hypothesized variance.")
else:
    print("\nFail to reject H0: No significant difference in population variance.")
```

```python
# Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

import numpy as np
import scipy.stats as stats

def z_test_proportions(x1, n1, x2, n2, alpha=0.05):
    """
    Perform a two-proportion Z-test.

    Parameters:
    x1, x2: Number of successes in each group
    n1, n2: Sample sizes of each group
    alpha: Significance level (default = 0.05)

    Returns:
    Z-score, p-value, and conclusion
    """
    # Sample proportions
    p1 = x1 / n1
    p2 = x2 / n2

    # Pooled proportion
    p = (x1 + x2) / (n1 + n2)

    # Standard error
    se = np.sqrt(p * (1 - p) * (1/n1 + 1/n2))

    # Z-score calculation
    z_score = (p1 - p2) / se

    # P-value for two-tailed test
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    # Critical Z-value for two-tailed test
    z_critical = stats.norm.ppf(1 - alpha/2)

    # Print results
    print("=== Two-Proportion Z-Test ===")
    print(f"Sample Proportions: p1 = {p1:.4f}, p2 = {p2:.4f}")
    print(f"Pooled Proportion: {p:.4f}")
    print(f"Z-Score: {z_score:.4f}")
    print(f"P-Value: {p_value:.4f}")
    print(f"Critical Z-Value: ¬±{z_critical:.4f}")

    # Decision Rule
    if abs(z_score) > z_critical or p_value < alpha:
        print("\nReject H0: Significant difference between proportions.")
    else:
        print("\nFail to reject H0: No significant difference between proportions.")

# Example Data: Two groups with different success rates
x1, n1 = 180, 400  # Group 1: 180 successes out of 400
x2, n2 = 150, 400  # Group 2: 150 successes out of 400

# Run Z-Test for proportions
z_test_proportions(x1, n1, x2, n2)
```

```python
# Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Generate random datasets
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10, Size=30
group2 = np.random.normal(loc=52, scale=15, size=30)  # Mean=52, Std=15, Size=30

# Compute sample variances
s1_squared = np.var(group1, ddof=1)  # Sample variance (unbiased)
s2_squared = np.var(group2, ddof=1)

# Assign larger variance to the numerator
if s1_squared > s2_squared:
    F_stat = s1_squared / s2_squared
    df1, df2 = len(group1) - 1, len(group2) - 1
else:
    F_stat = s2_squared / s1_squared
    df1, df2 = len(group2) - 1, len(group1) - 1

# Compute p-value (two-tailed test)
p_value = 2 * min(stats.f.cdf(F_stat, df1, df2), 1 - stats.f.cdf(F_stat, df1, df2))

# Print results
print("=== F-Test for Equality of Variances ===")
print(f"Sample Variances: Group1 = {s1_squared:.4f}, Group2 = {s2_squared:.4f}")
print(f"F-Statistic: {F_stat:.4f}")
print(f"Degrees of Freedom: df1 = {df1}, df2 = {df2}")
print(f"P-Value: {p_value:.4f}")

# Decision Rule (Œ± = 0.05)
alpha = 0.05
if p_value < alpha:
    print("\nReject H0: Significant difference in variances.")
else:
    print("\nFail to reject H0: No significant difference in variances.")

# Visualization of the F-Distribution
x = np.linspace(0, 5, 1000)
y = stats.f.pdf(x, df1, df2)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label=f"F-Distribution (df1={df1}, df2={df2})")
plt.axvline(F_stat, color='r', linestyle='--', label=f'F-Statistic = {F_stat:.2f}')
plt.fill_between(x, y, where=(x > stats.f.ppf(1-alpha/2, df1, df2)) | (x < stats.f.ppf(alpha/2, df1, df2)),
                 color='red', alpha=0.3, label='Rejection Regions (Œ±=0.05)')
plt.xlabel("F-Value")
plt.ylabel("Density")
plt.title("F-Test: Variance Comparison")
plt.legend()
plt.show()
```
